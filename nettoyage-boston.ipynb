Now we work on binary data

```python
binary_columns = ['host_is_superhost','host_has_profile_pic','host_identity_verified','is_location_exact','has_availability','requires_license','instant_bookable','is_business_travel_ready','require_guest_profile_picture','require_guest_phone_verification']
for i in binary_columns :
    categorical_df[i] = categorical_df[i].replace({'t': int(1), 'f': int(0)})
categorical_df
```

data that's actually Numerical :( like price but cuz of the dollar mark is classified as an object )

```python
num_data_in_cat=['host_response_rate','price','weekly_price','monthly_price','security_deposit','cleaning_fee','extra_people']
#host_response_rate is percentage we will remove the '%' mark then move the column to numericall data frame :
categorical_df['host_response_rate'] = categorical_df['host_response_rate'].str.replace('%', '').astype(float)
numerical_df = pd.concat([numerical_df,categorical_df['host_response_rate']],axis=1)
#rename the column to make the data more readable
numerical_df = numerical_df.rename(columns={'host_response_rate': 'percent_host_response_rate'})
#drop the column from the categorical dataframe
categorical_df = categorical_df.drop('host_response_rate', axis=1)
for i in range(1, len(num_data_in_cat)) :
    categorical_df[num_data_in_cat[i]] = categorical_df[num_data_in_cat[i]].str.replace(',', '') #if u dont do this you will get en error at values like 3,999.0
    categorical_df[num_data_in_cat[i]] = categorical_df[num_data_in_cat[i]].str.replace('$', '').astype(float)
    numerical_df = pd.concat([numerical_df,categorical_df[num_data_in_cat[i]]],axis=1)
    #rename the column to make the data more readable
    numerical_df = numerical_df.rename(columns={num_data_in_cat[i]: num_data_in_cat[i]+'_in$'})
    #drop the column from the categorical dataframe
    categorical_df = categorical_df.drop(num_data_in_cat[i], axis=1)

numerical_df
```

Nominal data to be unpacked, then later one hot encoded

```python
categorical_df['amenities'].unique()
categorical_df = categorical_df.drop('amenities',axis=1)
```

```python
categorical_df['host_verifications'].unique()
```

```python
import ast
unique_categories_host_verifications = []
for i  in list(categorical_df['host_verifications'].unique()) :
    my_list = ast.literal_eval(i)
    unique_categories_host_verifications.append(my_list)
#remove none vals :
unique_categories_host_verifications = [x for x in unique_categories_host_verifications if x is not None]
# now get the unique vals 
unique_categories_host_verifications0 = set(host_verification for row in unique_categories_host_verifications for host_verification in row)
unique_categories_host_verifications0
# for each type of verification we add a columns with 1 or 0 value
for i in unique_categories_host_verifications0 :
    categorical_df[i+'_host_verifications'] = categorical_df['host_verifications'].apply(lambda x: int(i in x))
categorical_df = categorical_df.drop('host_verifications',axis=1)
# unique_categories_host_verifications
```

```python
bed_types = categorical_df['bed_type'].unique()
bed_types
```

```python
#reate new columns to indicate bed type and drop the originale column
for i in bed_types :
    categorical_df['is_bed_type_'+i] = categorical_df['bed_type'].apply(lambda x: 1 if x == i else 0)
categorical_df = categorical_df.drop('bed_type',axis=1)
```

```python
property_types = categorical_df['property_type'].unique()
#we do the same to the property_type
for i in property_types :
    categorical_df['is_property_type_'+i] = categorical_df['property_type'].apply(lambda x: 1 if x == i else 0)
categorical_df = categorical_df.drop('property_type',axis=1)
```

```python
for i in property_types :
    print(categorical_df['is_property_type_'+i].value_counts())
```

```python
final_df = pd.merge(numerical_df,categorical_df,on='id')
```
